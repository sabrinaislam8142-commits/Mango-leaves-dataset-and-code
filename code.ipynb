```python
#  Load Pretrained YOLOv8s model

model = YOLO("yolov8s.pt")  # pretrained small model


# Train Model

model.train(
    data=f"{dataset_path}/data.yaml",  # dataset
    epochs=100,                        # number of epochs
    imgsz=640,                         # image size
    batch=16,                          # batch size (adjust based on GPU memory)
    project=results_path,              # save results in Google Drive
    name="train_run1"
)
```

```python
 #Validate the Model
model.val(
    data=f"{dataset_path}/data.yaml",
    project=results_path,
    name="val_run1"
)
```

```python
# Run Inference on Test Set

model.predict(
    source=f"{dataset_path}/images/test",
    project=results_path,
    name="predict_run1",
    save=True,
    save_txt=True,
    save_conf=True
)
```
